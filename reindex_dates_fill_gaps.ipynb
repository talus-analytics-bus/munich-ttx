```python
cum = pd.read_csv('/Users/trae/git_repositories/vaccines-ingest/data/IQVIA_place-of-vax/cumulative_iqvia_by_week_20201201.csv')

cum.head()
```

```python
cum_cols = cum.columns
```

```python
cum.drop('week_num', axis=1, inplace=True)
```

```python
cum.loc[((cum['state'] == 'Alabama') & (cum['season'] == '2019-2020')), :].set_index(['state', 'age', 'date']).ffill()
```

```python
cum_2019 = cum.loc[(cum['season'] == '2019-2020'), :]
cum_2019['date'] = pd.to_datetime(cum_2019['date'])
cum_2019.set_index(['state', 'age', 'date'], inplace=True)
new_index_2019 = pd.MultiIndex.from_product(cum_2019.index.levels)
cum_2019 = cum_2019.reindex(new_index_2019).ffill().reset_index()

cum_2020 = cum.loc[(cum['season'] == '2020-2021'), :]
cum_2020['date'] = pd.to_datetime(cum_2020['date'])
cum_2020.set_index(['state', 'age', 'date'], inplace=True)
new_index_2020 = pd.MultiIndex.from_product(cum_2020.index.levels)
cum_2020 = cum_2020.reindex(new_index_2020).ffill().reset_index()
```

```python
test = cum_2019.reset_index()

test.loc[(test['state'] == 'North Dakota') & (test['age'] == '18+'), :].tail(15)
```

```python
cum_reindexed = pd.concat([cum_2019, cum_2020])
```

```python
cum_reindexed.shape
```

```python
cum_reindexed['week_num'] = cum_reindexed['date'].dt.week
```

```python
cum_reindexed['week_num'].value_counts()
```

```python
cum_reindexed[cum_cols].to_csv('/Users/trae/git_repositories/vaccines-ingest/data/IQVIA_place-of-vax/cumulative_iqvia_ffill_20201201.csv', index=False)
```
